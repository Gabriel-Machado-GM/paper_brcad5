```python
import pandas as pd
import numpy as np
expert_id = pd.read_csv('../data/expert_label_identification.csv')
expert_info = pd.read_csv('../data/human_experts_data.csv')
```

```python
cases_per_expert = expert_id.groupby('volunteer_id').count().reset_index()
cases_per_expert
```

```python
years_experience_judiciary = cases_per_expert.merge(expert_info[['user_id', 'years_experience_judiciary']], how='inner', left_on='volunteer_id', right_on='user_id').copy()
years_experience_judiciary.drop('user_id', axis=1, inplace=True)
print(f'Weighted Average - Experience in Federal Courts: {np.average(years_experience_judiciary.years_experience_judiciary, weights=years_experience_judiciary.case_number)} years')
```

```python
years_higher_legal_education = cases_per_expert.merge(expert_info[['user_id', 'years_higher_legal_education']], how='inner', left_on='volunteer_id', right_on='user_id').copy()
years_higher_legal_education.drop('user_id', axis=1, inplace=True)
print(f'Weighted Average - Legal Higher Education: {np.average(years_higher_legal_education.years_higher_legal_education, weights=years_higher_legal_education.case_number)} years')
```

```python
years_experience_federal_small_claims_courts = cases_per_expert.merge(expert_info[['user_id', 'years_experience_federal_small_claims_courts']], how='inner', left_on='volunteer_id', right_on='user_id').copy()
years_experience_federal_small_claims_courts.drop('user_id', axis=1, inplace=True)
print(f'Weighted Average - Experience FSCC: {np.average(years_experience_federal_small_claims_courts.years_experience_federal_small_claims_courts, weights=years_experience_federal_small_claims_courts.case_number)} years')
```
